## ***Git Branch***

Com as branchs, vários desenvolvedores podem trabalhar em paralelo no mesmo projeto simultaneamente. Pode-se utilizar o comando git branch para criar, listar e excluir branchs.

Para visualizar as branchs: git branch or git branch --list

Para deletar uma branch: git branch -d nome-da-branch

Para subir a branch no github: gitpush -u remote; nome da branch


### ***Importância das branchs***

As branchs são diferentes pastas que os desenvolvedores podem trabalhar paralelamente sem interferir no código de outro desenvolvedor.

Por exemplo, trabalhando em apenas uma branch acontece o seguinte problema
image.png

Porém quando estamos trabalhando em branchs separadas, o código fica muito mais organizado como mostra a ilustração a seguir:

image.png


1. Com branches, vários desenvolvedores podem trabalhar em paralelo no mesmo projeto simultaneamente. Pode-se usar o comando git branch para **criar**, **listar** e **excluir** branches
2. Criando uma nova **branch:**
    1.  ***git branch <nome-do-branch>***  Este comando criará uma branch **localmente**
    2. Sendo **necessário** subir essa branch em um repositório remoto
3. Para **enviarmos** uma branch para o **repositório remoto**, você precisa usar o seguinte comando:
    1. ***git push -u <remote> <nome-da-branch>***
4.  Para **ver** as branches:
    1. ***git branch** ou **git branch --list***
5. **Deletando** uma branch:
    1. ***git branch -d <nome-da-branch>***


→ É possível perceber que existe apenas um **branch** para vários desenvolvedores. Se todo mundo for tentar *commitar (Entregável)* em uma mesma feature, no mesmo instante, irão ocorrer muitos conflitos, toda hora alguém vai acabar sobrescrevendo algo que já foi feito, e isso prejudica muito o fluxo de trabalho. 

→Para poucos ***devs*** até ficaria viável utilizar esse fluxo, pois não exige muita comunicação, mas e se tivermos 30/40/50 ***devs,*** como iremos manter um fluxo contínuo de comunicação e garantir que não exista conflito entre os entregáveis? Em **projetos** de desenvolvimento buscamos velocidade para conseguir desenvolver maior quantidade de *features (funcionalidades),* resolução de problemas e projetos no menor tempo possível.

Uma solução para os problemas é a separação de cada desenvolvedor em sua branch, e isso diminui a quantidade de conflitos críticos, pois a cada commit ou alteração em uma determinada branch acabam tendo um menor impacto caso ocorra um conflito, pois existe a possibilidade de retornamos a uma situação na qual a feature está correta, que sabemos o que estava acontecendo, e então como não ocorre a sobrescreção os riscos são menores.

1. O git **checkout**  é um dos comandos Git mais usados. Para trabalhar em uma branch, primeiro você precisa mudar para ela.
2. Usamos o Git checkout principalmente para alternanr de uma branch para outra. Também podemos usá-lo para verificar arquivos e commits:
    1. ***git checkout <nome-da-branch>*** Comando que direciona a branch, porque uma vez que ela está criada não significa que você vai trabalhar nessa branch.
    2. Há um atalho que permite criar e trocar para uma branch ao mesmo tempo, ***git checkout -b <nome-da-branch>***

## ***Git commit***

É o comando que rastrea o que você fez no arquivo e define o ponto de verificação no processo de desenvolvimento, é como se fosse a foto do arquivo.

1. ***git commit -m “mensagem explicando a mudança no código”***
2. É como se fosse um checkpoint, você pega o arquivo que fez ou modificou e dissesse “Até aqui isso foi feito”
3. Existem algumas boas práticas nesses *commits*, **algumas bibliotecas já trabalham com essas mudanças de maneira automática. Existe a *conventional commit* que acaba forçando você fazer esse *commit* em uma determinada estrutura, mas varia bastante de empresa para empresa, você as vezes acaba usando o nome do projeto e a mudança aplicada, ou nome do projeto e algum grau de dificuldade.
4. É um aspecto importante para manter a comunicação entre os códigos, ainda mais se você trabalha com sistema legado, sistemas descontinuados, esses *commits* permitem um maior entendimento do fluxo de construção da aplicação e em que pé está.

Comando: git commit -m "Mensagem"

## ***Git push***

Comando que envia e salva as confirmações no repositório remoto.

git push remote; nome da branch

Se sua branch foi criada recentemente, você precisa subir ela para o github.

1. O comando git ***push*** envia e salva suas confirmações no repositório remoto.
    1. ***git push <remote> <nome-do-branch>***, o nome da branch e remote são comandos opcionais, apenas ***git push*** já direciona aquela branch em questão para o repositório.
2. No entanto, se seu branch for criado recentemente, você também precisará fazer upload da branch com o seguinte comando:
    1. ***git push -u origin <nome-do-branch>***
    2. Isso deve ser feito quando o branch foi criado apenas localmente e não no repositório remoto, quando uma branch é criada no repositório remoto, com esse comando já fazemos isso. Quando criamos um repositório **dentro** do **github** ele já da uma sequência de comandos para executar esse processo. ***Origin*** é um apelido para a branch, mas você poderia ter usado ***batatinha***, entretanto recomenda-se o uso da palavra origin por ser padrão entre as empresas.

git push -u origin nome da branch

## ***Git merge***

1. ***git merge <nome-da-branch>***
2. Na branch principal você usa esse comando para juntar a branch desejada, e aí as alterações feitas nessa branch secundárias serão refletidas na sua principal.

## ***Git flow***

É a boa prática de separação das branchs, inicia-se com a branch develop e dela sai as releases que são as entregas do projeto, e depois ela volta para a branch develop, ao final do processo fazemos o merge entre a branch develop e a main.

1. Basicamente é um guia de boas práticas para utilização das branches dentro do github, existem:
    1. git workflow
    2. github flow
2. São diversas nomenclaturas mas o que realmente importa é lembrar que esse esquema de branch existem, você trabalhar com duas principais
    1. **Develop** → Dos desenvolvedores responsáveis pelo desenvolvimento
    2. **Main →** Branch que irá guiar o desenvolvimento
    3. As duas branches se conversam dentro das *releases*, fechando a entrega do projeto e eventuais incrementos futuros.
